```python
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split

# Sample sensor data
X = [[60, 0.5], [72, 0.7], [90, 2.1], [93, 2.5], [70, 1.0], [88, 2.0]]
y = [0, 0, 1, 1, 0, 1]

# Split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Model with RBF kernel
model = SVC(kernel='rbf')
model.fit(X_train, y_train)

# Predict a new reading
model.predict([[92, 2.3]])  # Output: [1] â†’ Predicts failure
```

```python
import matplotlib.pyplot as plt

# Mapping: Low = 0, High = 1
# Points: (Temperature, Vibration)
data_points = [
    (0, 0, 'No Failure'),
    (1, 0, 'No Failure'),
    (1, 1, 'Failure')
]

# Separate points by label
no_failure = [(x, y) for x, y, label in data_points if label == 'No Failure']
failure = [(x, y) for x, y, label in data_points if label == 'Failure']

# Plotting
plt.figure(figsize=(6, 6))
plt.scatter([x for x, y in no_failure], [y for x, y in no_failure],
            color='green', label='No Failure', s=100, edgecolors='k')
plt.scatter([x for x, y in failure], [y for x, y in failure],
            color='red', label='Failure', s=100, edgecolors='k')

plt.title('Non-linearly Separable Data')
plt.xlabel('Temperature (0 = Low, 1 = High)')
plt.ylabel('Vibration (0 = Low, 1 = High)')
plt.xticks([0, 1])
plt.yticks([0, 1])
plt.grid(True)
plt.legend()
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split

#Example data
X = [[68, 0.5], [90, 1.7], [87, 2.0], [65, 0.8], [80, 1.3]]
y = [0, 1, 1, 0, 1]

X_train, y_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

model.predict([[85,1.8]]) # output 1 : system Failer
```

```python
import matplotlib.pyplot as plt

importance = model.feature_importances_
plt.bar(['Temp', 'Vibration'], importance)
plt.title('Sensor Influence on Failure')
plt.show()
```
