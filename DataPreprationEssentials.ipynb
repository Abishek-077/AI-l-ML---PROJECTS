STEP 1: Remove or Fix Inconsistencies

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df =sns.load_dataset('titanic') #Built in titanic dataset
df.head()
```

```python
df.isnull().sum()
```

```python
df['age']= df['age'].fillna(df['age'].median())
df.dropna(subset=['deck'],inplace=True)
df['embarked'].fillna(df['embarked'].mode()[0],inplace=True)
df['embark_town'].fillna(df['embark_town'].mode()[0],inplace=True)
df.isnull().sum()
```

STEP 3: Remove Outliners

Remove Outliers

```python
Q1 = df['age'].quantile(0.25)
Q3 = df['age'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

#filter the outlinears
df = df[(df['age'] >= lower_bound) & (df['age'] <= upper_bound)]
df.head()
```

STEP 4 - Make sure what are the datatypes

```python
df.dtypes
```

Step 5: Correct Data Types

```python
df['class'] = df['class'].astype('category')
df['sex'] = df['sex'].astype('object')
df['embarked'] = df['embarked'].astype('category')
```

This helps with :
- memory optimization
- Categorical encoding


Encoding Categorical Features

```python
df = pd.get_dummies(df, columns=['sex', 'embarked', 'class'], drop_first=True)
```

STEP 7: Scale Numirical Features

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df[['age','fare']]=scaler.fit_transform(df[['Age','Fare']])
```

Step 8: Feature Engineering

Creates New Informative Features

```python
# family size
df['family_size'] = df['sibsp'] + df['parch'] + 1

#Alone or not
df['isAlone']= (df['family_size']==1).astype(int)
```
